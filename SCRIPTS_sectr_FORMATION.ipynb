```python
import requests
from bs4 import BeautifulSoup
import csv

# URL de la page à scraper
url = "https://www.formation-dz.com/annuaire-de-la-formation?start=10"

# Effectuer la requête HTTP et obtenir le contenu HTML de la page
response = requests.get(url)
html_content = response.content

# Parser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Créer une liste pour stocker les données extraites
data_list = []

# Trouver toutes les divs contenant les informations des établissements
divs = soup.find_all("div", class_="etablissement")

# Parcourir les divs et extraire les données des établissements
for div in divs:
    # Extraire les informations des établissements
    nom = div.find("h3").text.strip()
    description = div.find("div", class_="description").text.strip()
    adresse = div.find("div", class_="adresse").text.strip()
    telephone = div.find("div", class_="telephone").text.strip()

    # Ajouter les données extraites à la liste
    data = {
        "Nom": nom,
        "Description": description,
        "Adresse": adresse,
        "Téléphone": telephone
    }
    data_list.append(data)

# Enregistrer les données dans un fichier CSV
filename = "etablissements.csv"
with open(filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=["Nom", "Description", "Adresse", "Téléphone"])
    writer.writeheader()
    writer.writerows(data_list)

print("Les données ont été enregistrées dans le fichier", filename)
```
